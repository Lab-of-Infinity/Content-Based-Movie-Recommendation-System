## Scraping 10000 popular movies through Loop on 500 pages

```python
df = pd.DataFrame()
```

```python
from tqdm import tqdm
```

```python
for i in tqdm(range(501,1000)):
    response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=22b88585a3335e392c9ed6d3e45f1958&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])
    df = df.append(temp_df,ignore_index=True)
    time.sleep(3)
```

```python
df.head()
```

```python
df.to_csv('Scrap movies_500page.csv')
```
